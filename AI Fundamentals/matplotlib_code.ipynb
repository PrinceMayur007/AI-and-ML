```python
import numpy as np
import matplotlib.pyplot as plt     # plt is alias to matplotlib.pyplot

%matplotlib inline      # % as magic; tries to get the snapshot or visual output in same cell
```

##  Plots, Axes, labels, legends

```python
x_axis_values = np.linspace(1,10,20) # equally space between 1 and 10
x_axis_values
```

```python
y_axis_values = np.log(x_axis_values)
y_axis_values
```

```python
# line plot

plt.plot(x_axis_values, y_axis_values, 'blue')
plt.xlabel('X Axis Values')
plt.ylabel('Y Axis Values')
plt.title('Basic plot')
plt.show()
```

```python
# When we are running analysis, we might want to compare results.
# For that we might need to have the plots placed side by side.

# Pyplot API provides subject as an option.
plt.subplot(121)        # plt.subplot(nrows, ncols, plot_number i.e. 1,2,1 but we write 121, without commas)
plt.plot(x_axis_values, y_axis_values, 'blue')

plt.subplot(122)
plt.plot(x_axis_values, np.power(x_axis_values, 2), 'green'); # ; signifies that we finished plotting
```

So whatever we wre seeing the pyplot module API.

This API is good for quick plots, but Matplotlab recommends using object-oriented API as it provides more control and customization for the plots.

Most of these function also exist as methods in the matplotlib.axes.Axes class.

#### Introduction to the Object Oriented Method

The main idea in using the more formal Object Orientedmethod is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

```python
fig, ax = plt.subplots() # plt.subplots is equivalent to using fig = plt.figure() and then ax = fig.add_subplot(111)

ax.plot(x_axis_values, y_axis_values, 'blue')
ax.set_xlabel('X Axis Values') # Notice the use of set_ to begin methods
ax.set_ylabel('Y Axis Values')
ax.set_title('Set Title')
```

```python
# Create two subplots

f, (ax1, ax2) = plt.subplots(1,2, sharey = True) # share y axis value
ax1.plot(x_axis_values, y_axis_values)
ax1.set_title('Sharing Y axis')
ax2.scatter(x_axis_values, y_axis_values)
```

```python
# There are instances when we would have multiple lines in the same plot.

fig, ax = plt.subplots()
ax.plot(x_axis_values, np.power(x_axis_values,2), 'red',label = "Sqaure")
ax.plot(x_axis_values, np.power(x_axis_values,3), label = "Cube")
ax.legend() # shows legend; legend identify the line.
```

```python
# Scatter plots

plt.scatter(x_axis_values, y_axis_values)
```

```python
# Histogram
data_norm_dist = np.random.normal(0,100,100000) 
```

```python
# Box Plot
# List comprehension
box_data = [np.random.normal(0, std, 100) for std in range(1, 4)]
box_data

# rectangular box plot
plt.boxplot(box_data, vert = True, patch_artist = True);
```
